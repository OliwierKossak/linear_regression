```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#import data from csv file
exam_data = pd.read_csv('exam_mark.csv')

learning_minutes = exam_data['learning_minutes']
exam_points = exam_data['points']
attendance = exam_data['attendance']

#Splits data to independent and dependent
X = np.array(exam_data.iloc[:, :2].values)
y = np.array(exam_data.iloc[:, -1].values)

#Splits data for testing and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Standardization of data
sc_x = StandardScaler()
sc_x.fit(X)

X_std_train = sc_x.transform(X_train)
X_std_test = sc_x.transform(X_test)


regression = LinearRegression()

# training of linear model
regression.fit(X_std_train, y_train)

#model prediction
y_pred = regression.predict(X_std_test)

#creates scatter chart for student learning time in minutes and points from exam
plt.scatter(X[:, 1],y)
plt.xlabel('learning time in minutes')
plt.ylabel('points from exam')
plt.show()

# creates scatter chart for student attendance and points from exam
plt.scatter(X[:, 0],y, color='red')
plt.xlabel('student attendance')
plt.ylabel('points from exam')
plt.show()


#Creates a 3D scatter plot that displays the predicted points in relation to the standardized learning minutes and attendance variables
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_std_test[:, 0], X_std_test[:, 1], y_pred, c='r', marker='o')

ax.set_xlabel('Attendance (Std)')
ax.set_ylabel('Learning Minutes (Std)')
ax.set_zlabel('Predicted Points')
plt.show()

print('MAE training data: %.3f, test data: %.3f' % (
        mean_absolute_error(y_train, regression.predict(X_std_train)),
        mean_absolute_error(y_test, regression.predict(X_std_test))))

print('MSE training data: %.3f, test data: %.3f' % (
        mean_squared_error(y_train, regression.predict(X_std_train)),
        mean_squared_error(y_test, regression.predict(X_std_test))))

print('R^2 training data: %.3f, test data: %.3f' % (
        r2_score(y_train, regression.predict(X_std_train)),
        r2_score(y_test, regression.predict(X_std_test))))

```
